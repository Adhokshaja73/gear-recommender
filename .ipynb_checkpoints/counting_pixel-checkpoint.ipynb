```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from math import sqrt
```

```python
uphillImg = cv2.imread("road.jpg")
```

```python
pts = np.array([[100, 420], [200, 0],[420, 0],[500, 420]],
               np.int32) 
pts = pts.reshape((-1, 1, 2))
mask = np.zeros(uphillImg.shape[:2], dtype="uint8")
cv2.fillPoly(mask, [pts], (255, 255, 255))
maskedImg = np.dstack((uphillImg, mask))
```

```python
plt.imshow(maskedImg)
```

```python
samplePts = np.array([[200, 420], [200, 350],[350, 350],[350, 420]],
               np.int32) 
samplePts = samplePts.reshape((-1, 1, 2))
sampleMask = np.zeros(uphillImg.shape[:2], dtype="uint8")
cv2.fillPoly(sampleMask, [samplePts], (255, 255, 255))
sampleImg = np.dstack((uphillImg, sampleMask))
```

```python
plt.imshow(sampleImg)
```

```python
sampleImg.shape
```

```python
sampleMinX = 220
sampleMaxX = 350
sampleY = 350

avgColor = [0,0,0]
diff = [0,0,0]
for i in range(sampleMinX, sampleMaxX):
    currentPixel = uphillImg[sampleY][i]
    avgColor[0] = (avgColor[0] + currentPixel[0]) // 2
    
    avgColor[1] = (avgColor[1] + currentPixel[1]) // 2

    avgColor[2] = (avgColor[2] + currentPixel[2]) // 2

count = 0
for i in range(sampleMinX, sampleMaxX):
    currentPixel = uphillImg[sampleY][i]
    diff[0] = (diff[0] + (currentPixel[0] - avgColor[0])**2)
    diff[1] = (diff[1] + (currentPixel[1] - avgColor[1])**2)
    diff[2] = (diff[2] + (currentPixel[2] - avgColor[2])**2)
    count += 1

diff = [int(sqrt(i/count)) for i in diff]


def colorInRange(pixel):
    r = pixel[0]
    g = pixel[1]
    b = pixel[2]
    if(r in range(avgColor[0] - diff[0],avgColor[0] + diff[0])):
        if(g in range(avgColor[1] - diff[1],avgColor[1] + diff[1])):
            if(b in range(avgColor[2] - diff[2],avgColor[2] + diff[2])):
                return True
    return False

areaMinX = 150
areaMaxX = 400
areaMinY = 50
areaMaxY = 400
count = 0
pixelCount = 0
for i in range(areaMinX, areaMaxX):
    for j in range(areaMinY,areaMaxY):
        currentPixel = uphillImg[j][i]
        if(colorInRange(currentPixel)):
            count += 1
        pixelCount += 1
```
